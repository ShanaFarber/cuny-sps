# Week 1 - Hello Graph World Assignment

First, let's load the required packages and initialize a graph object.

```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.Graph()
```

Now, let's add the nodes from using `add_nodes_from()`.

```python
graph.add_nodes_from(["Andre", "Beverly", "Carol", "Diane", "Ed", "Fernando", "Garth", "Heather", "Ike", "Jane"])
```

Now, let's add the edges/relationships indicated in the assignment using `add_edges_from()`.

```python
graph.add_edges_from([("Diane", "Andre"), ("Diane", "Beverly"), ("Diane", "Ed"), ("Diane", "Garth"), ("Diane", "Fernando"), ("Diane", "Carol"),
                      ("Andre", "Fernando"), ("Andre", "Carol"), ("Andre", "Beverly"), 
                      ("Beverly", "Ed"), ("Beverly", "Garth"), 
                      ("Fernando", "Carol"), ("Fernando", "Garth"), ("Fernando", "Heather"),
                      ("Garth", "Ed"), ("Garth", "Heather"),
                      ("Heather", "Ike"),
                      ("Ike", "Jane")])

nx.draw(graph, with_labels = True)

graph.nodes()
```
