1. Import a “dirty” dataset from the internet into your environment. (Missing values, improper coding of columns, etc.)
2. Clean or “tidy” the data when loading into python using pandas.

Looking at the datasets we have, `df2` has a column for salary which should ideally be an integer datatype and there is also a column of just missing values. Looking at the `csv` for this dataframe, the empty column is due to an extra comma after each salary entry. Let's change the format of the salary column to be more usable by converting to integer and remove the extra column. The state column is also improperly named as "Sate."

```python
# clean df2
df2_cleaned = df2.rename(columns={"Sate ":"State"})

# change '2018 Mean Annual Wage' column to integer dtype
df2_cleaned['2018 Mean Annual Wage'] = df2_cleaned['2018 Mean Annual Wage'].apply(lambda x: int(x.replace(',', '')))

# remove empty column
df2_cleaned = df2_cleaned.drop('Unnamed: 2', axis=1)
```

Let's take a look now at the cleaned dataframe using `.head()`:

```python
df2_cleaned.head()
```

We have removed the extra column and fixed the naming of the `State` column. 

Let's take a look at the datatypes of this cleaned dataframe:

```python
df2_cleaned.dtypes
```

The datatype for `2018 Mean Annual Wage` is now an integer. 

We can also now see descriptive statistics for this column:

```python
df2_cleaned.describe()
```
