**Q1:** Create a class called BankAccount that has four attributes: bankname, firstname, lastname, and balance.

The default balance should be set to 0.

In addition, create ...
- A method called deposit() that allows the user to make deposits into their balance.
- A method called withdrawal() that allows the user to withdraw from their balance.
- Withdrawal may not exceed the available balance. Hint: consider a conditional argument in your withdrawal() method.
- Use the \_\_str__() method in order to display the bank name, owner name, and current balance.
- Make a series of deposits and withdrawals to test your class

```python
class BankAccount:
    # initiate
    def __init__(self, bankname, firstname, lastname, balance = 0):
        self.bankname = bankname
        self.firstname = firstname
        self.lastname = lastname
        self.balance = balance

    # method to withdraw from account
    def withdrawal(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Error! Insufficient funds.")

    # method to deposit to accont
    def deposit(self, amount):
        self.balance += amount

    def __str__(self):
        return f'Bank: {self.bankname}\nAccount Holder: {self.firstname} {self.lastname}\nBalance: ${self.balance:,}'

account = BankAccount("Chase", "Shoshana", "Farber", 35000)

print(account)

# withdraw money
print("\nWithdrawing $2,400")
account.withdrawal(2400)
print(account)

# deposit money
print("\nDepositing $5,000")
account.deposit(5000)
print(account)

# attempt to withdraw too much
# deposit money
print("\nAttempt to withdraw $45,000")
account.withdrawal(45000)
print(account)

# deposit money
print("\nDepositing $8,000")
account.deposit(8000)
print(account)

# withdraw money
print("\nWithdrawing $400")
account.withdrawal(400)
print(account)
```

**Q2:** Create a class Box that has attributes length and width that takes values for length and width upon construction (instantiation via the constructor).

In addition, createâ€¦
- A method called render() that prints out to the screen a box made with asterisks of length and width dimensions.
- A method called invert() that switches length and width with each other.
- Methods get_area() and get_perimeter() that return appropriate geometric calculations.
- A method called double() that doubles the size of the box. Hint: Pay attention to return value here.
- Implement \_\_eq__ so that two boxes can be compared using ==. Two boxes are equal if their respective lengths and widths are identical.
- A method print_dim() that prints to screen the length and width details of the box.
- A method get_dim() that returns a tuple containing the length and width of the box.
- A method combine() that takes another box as an argument and increases the length and width by the dimensions of the box passed in.
- A method get_hypot() that finds the length of the diagonal that cuts through the middle.

- Instantiate 3 boxes of dimensions 5,10 , 3,4 and 5,10 and assign to variables box1, box2 and box3 respectively.
- Print dimension info for each using print_dim().
- Evaluate if box1 == box2, and also evaluate if box1 == box3, print True or False to the screen accordingly.
- Combine box3 into box1 (i.e. box1.combine()).
- Double the size of box2.
- Combine box2 into box1.

```python
import math 

class Box:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    # renders box
    def render(self):
        for _ in range(self.width):
            print("*"*self.length)

    # invert length and width
    def invert(self):
        self.width, self.length = self.length, self.width

    # calculate area
    def get_area(self):
        return self.length*self.width

    # calculate perimeter
    def get_perimeter(self):
        return 2*(self.width+self.length)

    # double the size
    # I am unclear if "size" here is referring to area or dimensions 
    # therefore I am adding parameter "size" which defaults to dimensions
    # but can also be used to double the area 
    def double(self, size = "dimensions"):
        if size == "dimensions":
            return Box(2*self.length, 2*self.width)
        elif size == "area":
            return 2*self.length*self.width

    # check equal
    def __eq__(self, other):
        return self.width == other.width and self.length == other.length

    # print dimension
    def print_dim(self):
        print(f'{self.length}x{self.width}')

    # get tuple of dimensions
    def get_dim(self):
        return (self.length, self.width)

    # add dimension of another box
    def combine(self, other):
        self.length += other.length
        self.width += other.width
    
    # calculate hypotenuse
    def get_hypot(self):
        return math.sqrt(self.length**2+self.width**2)

    
# instantiate 3 boxes
box1 = Box(5,10)
box2 = Box(3,4)
box3 = Box(5,10)

# print dimentions
print("Box 1 Dimensions:", end=" "); box1.print_dim()
print("Box 2 Dimensions:", end=" ");box2.print_dim()
print("Box 3 Dimensions:", end=" ");box3.print_dim()

# evaluate if equal
print("Box 1 == Box 2:", end=" "); print(box1 == box2)
print("Box 1 == Box 3:", end=" "); print(box1 == box3)

# combined box1 and box3
box1.combine(box3)
print("Box 1 New Dimensions:", end=" "); box1.print_dim()

# double size of box2
doubled_box2_dims = box2.double()
doubled_box2_area = box2.double(size = "area")
print("Box 2 Doubled Dimensions:", end=" "); doubled_box2_dims.print_dim()
print(f"Box 2 Doubled Area: {doubled_box2_area}")

# combine box2 into box1
box1.combine(box2)
print("Box 1 New Dimensions:", end=" "); box1.print_dim()
```
